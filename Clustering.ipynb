# Clustering

1.What is unsupervised learning in the context of machine learning?
  - **Unsupervised Learning**: A type of machine learning where the algorithm identifiles patterns in data without labeled outputs (no target variable).

2.How Does K-Means clustering algorithm work?
  - **K-Means Clustering**: It partitions data into K clusters by minimizing within-cluster variance. It iteratively assigns points to the nearest centroid and updates centroidsd as the mean of assigned points.

3.Explain the concept of a dendrogram in hierachical clustering.
  - Dendrogram: A tree-like diagram in hierarchical clustering that shows the arrangement of clusters formed by merging or splitting them at various levels.

4.What is the main difference between K-Means and Hierarchical Clustering?
  - **K-Means vs. Hierarchical Clustering**: K-Means requires predefining the number of clusters and is efficient, while hierarchical clustering builds a nested cluster tree without needing the number of clusters upfront.

5.What are the advantages of DBSCAN over K-Means?
  - **Advantages of DBSCAN:**
     - Can find arbitrarily shaped clusters.
     - Doesn't require number of clusters beforehand.
     - Identifies noise/outliers.

6.When would you use Silhouette Score in clustering?
  - **Silhouette Score Usage:** Used to evaluate clustering quality by measuring how similar a point is to its own cluster compared to other clusters.

7.What are the limitations of Hierachical Clustering?
  - **Limitations of Hierarchical Clustering:**
    - Computationally intensive for large datasets.
    - Once a merge/split is made, it cannot be undone.
    - Sensitive to noise and outliers.

8.Why is feature scaling important in clustering algorithms like K-Means?
  - **Feature Scaling Importance:** Ensures all features contribute equally to distance calculations; otherwise, features with large scales dominate.

9.How does DBSCAN identify noise points?
  -** DBSCAN Noise Identification:** Points not within the minimum distance (epsilon) of a sufficient number of neighbors (minPts) are marked as noise.

10.Define inertia in the context of K-Means.
  - **Inertia in K-Means:** The sum of squared distances between data points and their assigned centroids; lower inertia indicates tighter clusters.

11.What is the elbow method in K-Means clustering?
  - **Elbow Method:** A technique to determine the optimal number of clusters by plotting inertia vs. number of clusters and finding the "elbow" point where inertia reduction slows.

12.Describe the concept of "density" in DBSCAN.
  - **Density in DBSCAN:** Defined by the number of points within a specified radius (epsilon). dense regions have more than minPts points in the radius.

13.Can hierarchical clustering be ised on categorical data?
  - **Hierarchical Clustering on Categorical Data**: Possible using appropriate distance measures (e.g., Hamming distance) and linkage methods suited for categorical data.

14.What does a negative silhouette Score indicate?
  -** Negative Silhouette Score:** Indicates poor clustering -- the point is likely assigned to the wrong cluster.

15.Explain the term "linkage criteria" in hierarchical clustering.
  - **Linkage Criteria:** Determines how the distance between clusters is computed in hierarchical clustering (e.g., single, complete, average, ward).

16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
  - **Poor Performance of K-Means:** It assumes equal-sized, spherical clusters, so it struggles with clusters of different shapes, sizes, or densities.

17.What are the core parameters in DBSCAN, and how do they influence clustering?
  - **DBSCAN Parameters:**
     - epsilon : Radius for neighborhood search.
     - minPts: Minimum number of points to from a dense region. They define cluster compactness and noise.

18.How does K-Means++ improve upon standard K-Means intialization?
  - **K-Means++:** Improves K-Means by initializing centroids more intelligently to spread them out, reducing chances of poor convergence.

19.What is agglomerative clustering?
  - **Agglomerative Clustering**: A bottom-up hierarchical clusttering approach where each point stats as its own cluster and clusters are iteratively merged.

20.What makes Silhouette Score a better metric that just inertia for model evaluation?
  - **Silhouette Score vs. Inertia:** Silhouette considers both intra-cluster conhesion and inter-cluster separations, giving a more balanced cluster quality measure than inertia alone.

21.Generator  synthetic data with 4 centers using make_blobs and aply K-Means clustering. Visualize using a scatter plot.

```python
# prompt: 21.Generator  synthetic data with 4 centers using make_blobs and aply K-Means clustering. Visualize using a scatter plot.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Visualize the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=50, c='lightgreen', marker='s', edgecolor='black', label='cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=50, c='orange', marker='o', edgecolor='black', label='cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=50, c='lightblue', marker='v', edgecolor='black', label='cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=50, c='red', marker='^', edgecolor='black', label='cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=250, marker='*', c='red', edgecolor='black', label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

29.Create syntetic data using make_blobs and evaluate silhouette scores for k - 2 to 5 . Display as a bar chart.

```python
# prompt: 9.Create syntetic data using make_blobs and evaluate silhouette scores for k - 2 to 5 . Display as a bar chart.

import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Evaluate silhouette scores for k = 2 to 5
silhouette_scores = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    labels = kmeans.labels_
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Display as bar chart
plt.bar(range(2, 6), silhouette_scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs. Number of Clusters")
plt.show()
```

31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
# prompt: 31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
y_pred = kmeans.fit_predict(X_scaled)

# Visualize with decision boundaries
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering with Decision Boundaries")
plt.legend()
plt.show()
```

44.Load the Iris dataset and plot the silhouette Coefficient for each sample after K-Means clustering.

```python
# prompt: 44.Load the Iris dataset and plot the silhouette Coefficient for each sample after K-Means clustering.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0) # Assuming 3 clusters, adjust as needed
cluster_labels = kmeans.fit_predict(X)

# Compute the silhouette scores for each sample
silhouette_avg = silhouette_score(X, cluster_labels)
sample_silhouette_values = silhouette_samples(X, cluster_labels)

# Create the silhouette plot
y_lower = 10
fig, ax = plt.subplots(figsize=(8, 6))
for i in range(3): # Assuming 3 clusters
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    color = cm.nipy_spectral(float(i) / 3)
    ax.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_values,
        facecolor=color,
        edgecolor=color,
        alpha=0.7,
    )
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

ax.set_title("The silhouette plot for the various clusters.")
ax.set_xlabel("The silhouette coefficient values")
ax.set_ylabel("Cluster label")

# The vertical line for average silhouette score of all the values
ax.axvline(x=silhouette_avg, color="red", linestyle="--")

ax.set_yticks([])  # Clear the yaxis labels / ticks
ax.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

plt.show()
```

```python
# prompt: 47.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate noisy blobs
X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X)

# Count cluster and noise points
cluster_counts = {}
noise_count = 0
for label in clusters:
    if label == -1:
        noise_count += 1
    else:
        cluster_counts[label] = cluster_counts.get(label, 0) + 1

print("Cluster Counts:", cluster_counts)
print("Noise Count:", noise_count)
```
